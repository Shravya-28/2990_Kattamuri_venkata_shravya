```python
train=pd.read_csv('/kaggle/input/practice-hackathon/Train.csv')
train.head()
```

```python
train.shape
```

```python
train['A'].unique()
```

```python
label=preprocessing.LabelEncoder()
train['A']=label.fit_transform(train['A'])
train['A'].unique()
```

```python
train.head()
```

```python
x=train.drop('Target',axis=1)
y=train['Target']
```

```python
x.head()
```

```python
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(y_train.value_counts())
```

```python
linear=LinearRegression()
linear.fit(x_train,y_train)
```

```python
y_new=linear.predict(x_train)
#y_new=y_new.astype(int)
y_new
```

```python
error=mean_squared_error(y_train,y_new)
error_rmse=np.sqrt(error)
error_rmse
```

```python
y1=linear.predict(x_val)
```

```python
error1=mean_squared_error(y_val,y1)
error1_rmse=np.sqrt(error1)
error1_rmse
```
